## Очистка описания вакансий

```python
import re
import nltk
import pymorphy2
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer

nltk.download('stopwords')

stop_words_ru = set(stopwords.words('russian'))
stop_words_en = set(stopwords.words('english'))
stop_words = stop_words_ru.union(stop_words_en)

morph = pymorphy2.MorphAnalyzer()

def clean_text(text):
    # Удаление HTML-тегов
    text = re.sub(r"<.*?>", " ", text)

    # Удаление лишних пробелов
    text = re.sub(r" +", " ", text)

    # Удаление всех символов, которые не являются буквами, цифрами или пробелами
    text = re.sub(r"[^\w\s]", " ", text)

    text = re.sub(r'(:\w+:)', '', text)  # Удаление эмодзи

    # Токенизация текста
    tokens = WhitespaceTokenizer().tokenize(text)

    # Приведение к нижнему регистру и удаление стоп-слов
    tokens = [token.lower() for token in tokens if token.lower() not in stop_words]

    # Лемматизация токенов
    tokens = [morph.parse(token)[0].normal_form for token in tokens]

    return tokens
```

```python
from datetime import datetime

def process_resumes(entry, failed_resume_flag, rabota, vacancy_uuid):
    # Смотрим, работаем с failed_resumes или confirmed_resumes
    resumes_type = 'failed_resumes' if failed_resume_flag else 'confirmed_resumes'
    now = datetime.now()

    # Проходим по каждому резюме в данной категории
    for item in entry[resumes_type]:
        uuid = item['uuid']
        num_educ = 0 # Количество educations
        h_educ = 0 # Количество высших
        age = 0  # Устанавливаем значение age в 0
        educ_year = None
        duration_max = 0

        # Смотрим количество образований
        if 'educationItem' in item:
          educ_min = 2500
          educ_year = None
          for education in item['educationItem']:
            num_educ += 1
            if education['education_level'] == "Высшее":
              h_educ += 1
              educ_year = education['year']
              if educ_year is not None and (educ_min is None or int(educ_min) > int(educ_year)):
                educ_min = int(educ_year)


        # Достаем общее время работы человека + начало его карьеры и конец
        if 'experienceItem' in item:
            end = item['experienceItem'][0].get('ends', "2024-02-25")

            for experience in item['experienceItem']:
                starts = experience.get('starts', None)
                ends = experience.get('ends', None)

                starts_date = datetime.strptime(starts, "%Y-%m-%d") if starts else None
                ends_date = datetime.strptime(ends, "%Y-%m-%d") if ends else datetime.strptime("2024-02-25", "%Y-%m-%d")

                # Расчет разницы в днях
                duration = (ends_date - starts_date).days if starts_date and ends_date else 0
                duration_max += int(duration)

            rabota.loc[(rabota['ends'].isnull()) & (rabota['starts'].notnull()), 'ends'] = '2024-02-25'
        else:
            starts = None
            end = None
            duration_max = 0

        # Смотрим количество лет
        if item['birth_date'] is not None:
            birth_date_str = item['birth_date']
            birth_date = datetime.strptime(birth_date_str, "%Y-%m-%d")
            today = datetime.now()
            age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
        # Количество лет людей, у которых не указан возраст, но есть высшее или начало работы
        elif item['birth_date'] is None:
          if educ_year is not None:
            age = now.year - educ_min + 21
          else:
            age = now.year - starts_date.year + 20

        rabota = rabota.append({'uuid': uuid, 'vacancy_uuid': vacancy_uuid, 'failed_resume': int(failed_resume_flag), 'starts': starts, 'ends': end, 'duration': duration_max, 'age': age, 'num_educ': num_educ, 'h_educ': h_educ}, ignore_index=True)

    return rabota

# Создаем DataFrame
rabota = pd.DataFrame(columns=['uuid', 'vacancy_uuid', 'failed_resume', 'starts', 'ends', 'duration', 'age', 'num_educ', 'h_educ'])

# Проходим по всем элементам в data
for entry in data3:
    # Для каждой вакансии
    vacancy_uuid = entry['vacancy']['uuid']

    # Для failed_resumes
    rabota = process_resumes(entry, True, rabota, vacancy_uuid)

    # Для confirmed_resumes
    rabota = process_resumes(entry, False, rabota, vacancy_uuid)
```
